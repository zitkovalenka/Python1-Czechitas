# Histogram platů

```python
import requests
import pandas
import matplotlib.pyplot as plt

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/platy_2021_02.csv")
open("ukol-12_platy_2021_02.csv", "wb").write(r.content)
```

```python
platy = pandas.read_csv("ukol-12_platy_2021_02.csv")
platy
```

```python
platy["plat"].hist(
    bins= 6,
    range = (30000,60000),
     rwidth = 0.9
)
plt.title("Plat v Kč\n\n", fontweight ="bold")
plt.show()
```

```python

zam_praha = pandas.read_csv("ukol-10_zam_praha.csv") 
zam_plzen = pandas.read_csv("ukol-10_zam_plzeň.csv") 
zam_liberec = pandas.read_csv("ukol-10_zam_liberec.csv") 

#  Vytvoř novou tabulku zamestnanci a ulož do ní informace o všech zaměstnancích.
# vytvořila jsem si nový index pro novou tabulku (ignore index), aby byl 0-n

zam_praha["mesto"] = "Praha"
zam_plzen["mesto"] = "Plzeň"
zam_liberec["mesto"] = "Liberec"

zamci = pandas.concat([zam_praha,zam_plzen,zam_liberec], ignore_index = True)
zamci
```

```python
platy_zamci = pandas.merge(zamci,platy, on = ["cislo_zamestnance"], how="left")
platy_zamci
# platy_zamci.shape
```

```python
platy_zamci.hist(
    by="mesto",
    # bins = [30000,35000,40000,45000,50000,55000,60000],
    bins = 6,
    range = (30000,60000),
    xrot = 40,
    # jak zde mám vytvořit název osy X?
    )

fig, ax = plt.subplots()    # jednotka 15 a 7 je v palcích
print(type(fig))
print(type(ax))

ax.set_title("Platy v Kč", fontweight="bold", fontsize = 24)    # proč se mi objeví graf zvlášť jen s "rámečkem"
# ax.set_xlabel("Platy v Kč", color="darkgrey", fontsize = 12)  # proč se mi objeví graf zvlášť jen s "rámečkem"

plt.savefig('ukol-12_platy.png')   # nutné volat před plt.show(), aby se obrázek uložil s grafem, ne prázdný 
plt.savefig('ukol-12_platy.jpg')   # nutné volat před plt.show(), aby se obrázek uložil s grafem, ne prázdný 
plt.show()
```

# Teplota ve městech

```python
import requests

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/temperature.csv")
open("ukol-12_temperature.csv", "wb").write(r.content)

temperature = pandas.read_csv("ukol-12_temperature.csv")
temperature
```

```python
# Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.

temp_3_city = temperature[temperature["City"].isin(["Helsinki", "Miami Beach", "Tokyo"])]
temp_3_city

#  nebo
temp_3_city = temperature.query("City == 'Helsinki' or City == 'Miami Beach' or City == 'Tokyo'")
temp_3_city
```

```python
# vypíšu všechny záznamy splňující podmínku, ne jen prvních 5 a posledních 5 záznamů.

# Takhle zmenis chovani pandas obecne pro cely script:
# pandas.set_option("display.max_rows", None, "display.max_columns", None)

# Tohle jen pro to jedno konkretni volani:
with pandas.option_context('display.max_rows', None, 'display.max_columns', None):  
    print(temp_3_city)

# Bude to fungovat v terminalu, kdyz pouzijes .py soubor.

# Pro Jupyter dostanes hlasku: Output exceeds the size limit. Open the full output data in a text editor
# Kliknutim na limit muzes upravit, kolik radek bude vracet (porad potrebujes mit spravne nastaven i ten pandas.
# Kliknutim na text editor otevres cely dataframe v novem okne.
# Vice moznovati nastaveni najdes v dokumentaci: https://pandas.pydata.org/pandas-docs/stable/user_guide/options.html
```

```python
# Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.

temp_3_city.plot(
    column = "AvgTemperature",
    by="City", 
    kind="box", 
    whis=[0,100]
)
plt.savefig('ukol-12-temp_3_city.png')   # nutné volat před plt.show(), aby se obrázek uložil s grafem, ne prázdný
plt.savefig('ukol-12-temp_3_city.jpg')   # nutné volat před plt.show(), aby se obrázek uložil s grafem, ne prázdný 
plt.show()

#  nevčetla jsem to z https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html . Vůbec tam argument "column", ani "by" není. Jen jsem to zkusila, protože jsem nevěděla, jak na to.
```
