```python
# Načti data o zaměstnancích z CSV souborů do tabulek (DataFrame). 

import requests

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/zam_praha.csv")
open("ukol-10_zam_praha.csv", "wb").write(r.content)

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/zam_plzeň.csv")
open("ukol-10_zam_plzeň.csv", "wb").write(r.content)

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/zam_liberec.csv")
open("ukol-10_zam_liberec.csv", "wb").write(r.content)

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/platy_2021_02.csv")
open("ukol-10_platy_2021_02.csv", "wb").write(r.content)


import pandas

zam_praha = pandas.read_csv("ukol-10_zam_praha.csv") 
zam_plzen = pandas.read_csv("ukol-10_zam_plzeň.csv") 
zam_liberec = pandas.read_csv("ukol-10_zam_liberec.csv") 
platy = pandas.read_csv("ukol-10_platy_2021_02.csv") 
```

```python
# Ke každé tabulce přidej nový sloupec mesto, které bude obsahovat informaci o tom, ve kterém městě zaměstnanec pracuje.

zam_praha["mesto"] = "Praha"
zam_plzen["mesto"] = "Plzeň"
zam_liberec["mesto"] = "Liberec"
```

```python
# zam_praha
# zam_plzen
zam_liberec
```

```python
#  Vytvoř novou tabulku zamestnanci a ulož do ní informace o všech zaměstnancích.
# vytvořila jsem si nový index pro novou tabulku (ignore index), aby byl 0-n

zamci = pandas.concat([zam_praha,zam_plzen,zam_liberec], ignore_index = True)
zamci
```

```python
# Ze souboru platy_2021_02.csv načti platy zaměstnanců za únor 2021. Propoj tabulku (operace join) s platy a tabulku se zaměstnanci pomocí sloupce cislo_zamestnance.
# načtení hotovo viz výše

# platy_zamci = pandas.merge(zamci,platy, on = ["cislo_zamestnance"])
platy_zamci = pandas.merge(zamci,platy, on = ["cislo_zamestnance"], how="left")
platy_zamci.shape
```

```python
platy_mesto_prum = platy_zamci.groupby("mesto")[["plat"]].mean()
platy_mesto_prum
```

```python
# V rámci úspory se IT oddělení rozhodlo prověřit licence přidělené zaměstnancům, kteří ve firmě již nepracují. Vytvoř samostatnou tabulku, která obsahuje jména zaměstnanců, 
# kteří ve firmě již nepracují. Tabulku ulož do souboru CSV.

# platy_zamci["plat"].isnull()

import numpy

platy_zamci["je_zamec"] = numpy.where(platy_zamci["plat"].isnull(),"Ne","Ano")
platy_zamci
```

Propoj tabulku s výkazy s tabulkou se zaměstnanci, kterou jsi vytvořil(a) v předchozím cvičení.

Následně proveď statistiku vykázaných hodin za jednotlivé kanceláře, tj. spočítej celkový počet hodin vykázaný zaměstnanci jednotlivých kanceláří na projekty daného zákazníka.

```python
# import requests

r = requests.get("https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/vykazy.csv")
open("ukol-10_vykazy.csv", "wb").write(r.content)
```

```python
# import pandas

vykazy = pandas.read_csv("ukol-10_vykazy.csv") 
vykazy
```

```python
# Proveď agregaci a zjisti celkový počet vykázaných hodin za jednotlivé projekty.

vykazy.groupby("project")[["project","hours"]].sum()
```

```python
# Propoj tabulku s výkazy s tabulkou se zaměstnanci, kterou jsi vytvořil(a) v předchozím cvičení.

# přejmenování sloupce z emloyee_id na employee_id
vykazy = vykazy.rename(columns={"emloyee_id":"employee_id"})
```

```python
# spojení dvou dataframe s jiným názvem klíčových sloupců (obsahují ale identický klíč)
zamci_hod_vykazy = pandas.merge(zamci, vykazy, left_on = "cislo_zamestnance", right_on = "employee_id")¨


# Kdyz vybiras jen nejake sloupce, musis vzdy vybrat ty, pres ktere to spojujes (a jakekoliv dalsi k tomu). Zde ten spojovací je employee_id, které mi chybělo
zamci_hod_vykazy = pandas.merge(zamci[["jmeno","prijimeni","cislo_zamestnance", "mesto"]], vykazy[["hours","project", "employee_id"]], left_on = "cislo_zamestnance", right_on = "employee_id")
zamci_hod_vykazy
```

```python
# Následně proveď statistiku vykázaných hodin za jednotlivé kanceláře, tj. spočítej celkový počet hodin vykázaný zaměstnanci jednotlivých kanceláří na projekty daného zákazníka.

zamci_hod_vykazy.groupby(["mesto","project"])[["hours"]].sum()

# nebo
zamci_hod_vykazy.groupby(["mesto","project"]).agg({"hours":["sum"]})
```
